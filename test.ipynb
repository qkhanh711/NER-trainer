```python
def readfile(filename):
    '''
    read file
    '''
    f = open(filename)
    data = []
    sentence = []
    label= []
    for line in f:
        if len(line)==0 or line.startswith('-DOCSTART') or line[0]=="\n":
            if len(sentence) > 0:
                data.append((sentence,label))
                sentence = []
                label = []
            continue
        splits = line.split(' ')
        sentence.append(splits[0])
        label.append(splits[-1][:-1])

    if len(sentence) >0:
        data.append((sentence,label))
        sentence = []
        label = []
    return data

readfile("../BERT-NER/data/train.txt")
# print(len(readfile("../BERT-NER/data/train.txt")))
```

```python
import random


def generate_random_task(function_list, status_list, time_list):
  """Generates a random task input and corresponding output dictionary.

  Args:
      function_list: List of supported task functions.
      status_list: List of possible task statuses.
      time_list: List of possible due dates for tasks.

  Returns:
      A dictionary containing the generated input string and output dictionary.
  """
  # Select a random function
  function = random.choice(function_list)

  # Generate task name with a random number
  rand_int = random.randint(100, 999)
  task_name = f"task{rand_int}"

  output = {"function": function}

  if function in ("add task", "modify task", "delete task", "create task", "edit task"):
    if function in ("add task", "create task"):
      status = "new"
    elif function in ("modify task", "edit task"):
      status = random.choice(status_list)
    else:
      status = random.choice(["done", "close"])
    time = random.choice(time_list)
    output["parameters"] = {"task": task_name, "status": status, "time": time}

  elif function in ("list tasks", "show task"):
    status = random.choice(status_list) if random.random() < 0.5 else None
    time = "now"
    output["parameters"] = {"task": task_name, "status": status}
  elif function in ("help", "remind", "alarm"):
    time = random.choice(time_list)
    output["parameters"] = {"task": task_name, "time": time}
  else:
    raise ValueError(f"Unsupported function: {function}")

  input_string = f"{function} {activites[rand_int % len(activites)]}"

  if time and function not in ("list tasks", "show task"):
    input_string += f" on {time}"
  if status and function not in ("list tasks", "show task"):
    input_string += f" ({status})"

  input = {
    "inputID": task_name,
    "request": input_string,
  }

  return {
          "input": input, 
          "output": output
          }


# Example usage
if __name__ == "__main__":
  import json
  with open("random_input.json", "w") as f:
    for i in range(100):
      try:
        random_task = generate_random_task(function_list, status_list, time_list)
        f.write(json.dumps(random_task) + "\n")
        print(f"Input: {random_task['input']}")
        print(f"Output: {random_task['output']}")
      except:
        print("Error: Unsupported function")
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("Babelscape/wikineural-multilingual-ner")
model = AutoModelForTokenClassification.from_pretrained("Babelscape/wikineural-multilingual-ner")

nlp = pipeline("ner", model=model, tokenizer=tokenizer, grouped_entities=True)
```

```python
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)
```
